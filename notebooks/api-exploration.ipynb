```python
import json 
from openaq import OpenAQ
```

Grab api-key

```python
with open('../secrets.json') as f:
    secrets = json.load(f)
    f.close()
    
openaq_key= OpenAQ(secrets['openaq-api-key'])
```

Open up a client to see if we get a successful connection

```python
client = OpenAQ(secrets['openaq-api-key'])

client
```

The regions considered include Africa and Arab countries

```python
uk_bbox_monitoring_locations = client.locations.list(
    bbox=(-2.430293,-19.160156,35.189669,33.750000),
    limit=1000   
)

uk_bbox_monitoring_locations.meta
```

print out the sensors location

```python
uk_bbox_monitoring_locations.results[0]
```

```python
uk_bbox_monitoring_locations.results[0].__dict__
```

Grab sensors location ID and name

```python
locations_info = {} 

for location in uk_bbox_monitoring_locations.results:
    locations_info[location.id] = location.name
    
locations_info
```

```python
len(locations_info)
```

```python
locations_info = {
    k: v
    for k, v in locations_info.items()
    if isinstance(v, str) and v.isascii()
}
locations_info 
```

```python
len(locations_info)
```

Save locations_info as json to use later in the project

```python
with open('../locations_info.json', 'w', encoding='utf-8') as f:
    json.dump(locations_info, f, ensure_ascii=False, indent=4)
    f.close()
```
